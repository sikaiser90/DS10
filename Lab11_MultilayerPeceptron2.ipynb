```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers
import numpy as np

#1) ISIR 데이터셋
X,y = load_iris().data, load_iris().target

#2) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size = 1/3)

#3) Perceptron model 을 만들겠습니다
model = models.Sequential()
model.add(layers.Dense(units = 100, activation = 'linear', input_shape = X_train[1].shape))
model.add(layers.Dense(units = 100, activation = 'linear'))
model.add(layers.Dense(1, activation = 'linear'))

#4) 학습 파라미터 설정
model.compile(optimizer = 'sgd',
             loss = 'mse',
             metrics = 'acc')

#5) 학습 시작
model.fit(X_train, y_train, epochs = 1000, batch_size=1)

Xtest_loss,Xtest_acc = model.evaluate(X_test,y_test)
print(Xtest_acc)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from tensorflow.keras import models, layers, regularizers

x = np.arange(-20, 20, 0.1) 
y = np.sin(2*x)+np.power(x,2)*np.log(np.power(np.cos(x),2)+2)

plt.plot(x,y)
plt.show()
```

```python
model = models.Sequential()

model.add(layers.Dense(100, activation = "relu", input_shape=(1,)))
model.add(layers.Dense(100, activation = "relu")
model.add(layers.Dense(1,activation='linear'))
          
model.compile(optimizer='adam',loss='mean_squared_error',metrics = ['acc'])

model.fit(x,y,epochs = 5000,batch_size=4096*2, validation_split = 0.2)

ytest = model.predict(x)
plt.plot(x,y)
plt.plot(x,ytest)

plt.show()
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt 

batch_size = 4096
num_classes = 10
epochs = 50

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
plt.figure()
plt.imshow(x_train[0])
plt.show()
```

```python
x_train = x_train.reshape(60000, x_train.shape[1]*x_train.shape[2])
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
print(y_test[0])
```

```python
import tensorflow.keras

# convert class vectors to binary class matrices
y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes)
y_test = tensorflow.keras.utils.to_categorical(y_test, num_classes)

print(y_test[0])
```

```python
# model generation

model = Sequential()
## model add
model.add(layers.Dense(units = 1028, activation = 'relu',input_shape = (784,)))
model.add(layers.Dense(units = 2048, activation = 'relu'))
model.add(layers.Dense(units = 128, activation = 'relu'))
model.add(layers.Dense(units = num_classes, activation = 'softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size*8,
          epochs=10,
          verbose=1)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers
import numpy as np
import tensorflow.keras
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
#1) ISIR 데이터셋

X,y = load_iris().data, load_iris().target

#2) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size = 1/3)

#2-1) X_train, y_train -> PCA / LDA 둘중에 하나 해보시고, 그걸로 MLP를 돌ㄹ세요
lda = LDA()
lda.fit(X_train, y_train)
X_train, X_test = lda.transform(X_train),lda.transform(X_test)

num_classes = 3
# convert class vectors to binary class matrices
# y_train = tensorflow.keras.utils.to_categorical(y_train, num_classes)
# y_test = tensorflow.keras.utils.to_categorical(y_test, num_classes)

#3) MLPerceptron model 을 만들겠습니다
model = models.Sequential()
model.add(layers.Dense(units = 8, activation = 'relu', input_shape = X_train[1].shape))
model.add(layers.Dense(8, activation = 'relu'))
model.add(layers.Dense(8, activation = 'relu'))
model.add(layers.Dense(3, activation = 'softmax'))

#4) 학습 파라미터 설정
model.compile(optimizer = 'sgd',
             loss = 'sparse_categorical_crossentropy',
             metrics = 'acc')

#5) 학습 시작
model.fit(X_train, y_train, epochs = 1000, batch_size=100)

Xtest_loss,Xtest_acc = model.evaluate(X_test,y_test)
print(Xtest_acc)
np.argmax(model.predict(X_test), axis = 1)
```
