```python
# adaline using keras
# XOR 안되는 이유 확인


from tensorflow.keras import models, layers, optimizers, losses
import numpy as np

X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

T = np.array([0,1,1,0])

model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='sigmoid',input_shape=(2,)))

model.compile(optimizer='adam',
             loss="binary_crossentropy",
             metrics = ['acc'])

model.fit(X,T,epochs = 10000, batch_size = 1, verbose = 0)
```

```python
# test
Xtest_loss,Xtest_acc = model.evaluate(X,T)
print(Xtest_acc)
Ttest = model.predict(X)
print(Ttest)
```

```python
# adaline using keras
from tensorflow.keras import models, layers, optimizers, losses
import numpy as np

X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

T = np.array([0,0,0,1])

model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(2,)))
#units : layer에 있는 node의 개수

model.compile(optimizer='sgd',
             loss="mse",
             metrics = ['acc'])

model.fit(X,T,epochs = 10000, batch_size = 1, verbose = 0)
```

```python
# test
Xtest_loss,Xtest_acc = model.evaluate(X,T)
print(Xtest_acc)
Ttest = model.predict(X)
print(Ttest)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(y_test)
```

```python
from tensorflow.keras import models, layers, optimizers, losses

model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(4,)))

model.compile(optimizer='sgd',
             loss="mse",
             metrics = ['acc'])

model.fit(X_train,y_train,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# test
Xtest_loss, Xtest_acc = model.evaluate(X_test,y_test)
print(Xtest_acc)

Ttest = model.predict(X_test)
print(Ttest)
```

## 정답

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers
import numpy as np

X, y = load_iris().data, load_iris().target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3)

model = models.Sequential()
model.add(layers.Dense(units=1, activation = 'linear', input_shape = X_train[0].shape))

model.compile(optimizer='sgd', loss='mse', metrics = 'acc')

model.fit(X_train, y_train, epochs = 1000)
```

```python
Xtest_loss, Xtest_acc = model.evaluate(X_test,y_test)
print(Xtest_acc)

Ttest = model.predict(X_test)
print(Ttest)
```
